```python
# Task 1
# Sentiment Analysis
# Enter User Review : The movie was disaster
# Output : negative üòû
```

```python
# Task 1:
# Sentiment Analysis
# Example:
# Enter User Review : The movie was disaster
# Output : negative üòû

# Get user input and remove extra spaces
review = input("Enter User Review: ").strip()

# Prompt sent to OpenAI to classify sentiment
# The instruction forces the model to return only ONE word
prompt = f"""
Classify the sentiment of this review as Positive, Negative, or Neutral.
Return only one word.

Review: "{review}"
"""

# Send request to OpenAI Responses API
response = client.responses.create(
    model="gpt-5.2",
    input=prompt
)

# Safely extract the text output
# output_text is a shortcut but may not always exist
if hasattr(response, "output_text") and response.output_text:
    sentiment = response.output_text.strip().lower()
else:
    sentiment = response.output[0].content[0].text.strip().lower()

# Map sentiment label to emoji
emoji_map = {
    "positive": "üòä",
    "negative": "üòû",
    "neutral": "üòê"
}

# Display final output
print("Output:", sentiment, emoji_map.get(sentiment, ""))
```

```python
# Original user prompt
user_prompt = "i am a beginner , needs to learn neural networks , in a beginner friendly tone"

# Enhanced (professional) prompt
pro_prompt = (
    "Explain neural networks to me as a complete beginner, "
    "using simple language, clear examples, and a friendly, easy-to-understand tone."
)

# Send the enhanced prompt to OpenAI
response = client.responses.create(
    model="gpt-5.2",
    input=pro_prompt
)

# Print the response
print(response.output_text)

```

```python
# Input text and target language
source_language = "Malay"
target_language = "English"
text_to_translate = "saya suka kek coklat dengan pistachio cair yang mengandungi kunafa rangup"

# Prompt for translation
prompt = (
    f"Translate the following text from {source_language} to {target_language}. "
    f"Only return the translated word.\n\n"
    f"Text: {text_to_translate}"
)

# Send request to OpenAI
response = client.responses.create(
    model="gpt-5.2",
    input=prompt
)

# Print translated output
print(response.output_text)
```
